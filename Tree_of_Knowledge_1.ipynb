```python
import stanza
from nltk import tokenize
```

```python
import pandas as pd
from tqdm import tqdm
import numpy as np
import torch
import gc
```

```python
pipe = stanza.Pipeline("en", processors="tokenize,coref,mwt,pos,ner")
```

```python
import os
files = []
for filename in os.listdir():
    if filename.endswith(".txt"):
        files.append(filename)
files
```

```python
gc.collect()
empty_cache()
```

```python
t = tqdm(total=len(files), leave=False)
papers = []
antecedents_in_papers = []

for filename in files:
    print(filename)
    with open(filename, 'r', encoding="utf8") as f:
        article = f.read()
    text = article

    gc.collect()
    empty_cache()
    try:
        result = pipe(text)
    except Exception as e:
        print(filename, e)

    dicts = result.to_dict()

    resd = {}
    for a in dicts:
        for d in a:
            try:
                if d['coref_chains']:
                    if (d['ner'] == 'O' or 'S-ORG') and (d['upos'] not in ('PUNCT', 'PROPN', 'NUM', 'CCONJ', 'PRON')) and ('PronType=Art' not in d['feats']):
                        if d['coref_chains'][0].to_json()['representative_text'] in resd:
                            resd[d['coref_chains'][0].to_json()['representative_text']].append([d['text'], d['start_char'], d['end_char']])
                        else:
                            resd[d['coref_chains'][0].to_json()['representative_text']] = [[d['text'], d['start_char'], d['end_char']]]
            except Exception as e:
                continue

    # мердж последовательных
    for antecedent in resd:
        for i in range(0,len(resd[antecedent])-1):
            if resd[antecedent][i+1][1] - resd[antecedent][i][2] <= 1:
                resd[antecedent][i+1][1] = resd[antecedent][i][1]

    for antecedent in resd:
        resa = {idx[1]: [] for idx in resd[antecedent]}
        resmax = {idx[1]: [] for idx in resd[antecedent]}

        for idx in resd[antecedent]:
            resa[idx[1]].append(idx[0])
            resmax[idx[1]].append(idx[2])

        resd[antecedent] = []
        for ana in [[' '.join(resa[x]), x, max(resmax[x])] for x in resa]:
            if antecedent != ana[0]:
                resd[antecedent].append(ana)

    for x in resd.keys():
        if resd[x] != []:
            antecedents_in_papers.append(resd[x][0][0])
            papers.append(filename)
    #antecedents_in_papers += [resd[x][0][0] for x in resd.keys()]
    #papers += [filename]*len(resd.keys())

    t.update(1)
```

```python
df_out = pd.DataFrame().from_dict({'papers':papers, 'antecedents_in_papers':antecedents_in_papers})
```
